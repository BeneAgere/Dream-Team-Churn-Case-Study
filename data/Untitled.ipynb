```python
df['avg_rating_by_driver'] = df['avg_rating_by_driver'].fillna(df['avg_rating_by_driver'].sum()*1.0/len(df['avg_rating_by_driver']))
```

```python
df['avg_rating_of_driver'] = df['avg_rating_of_driver'].fillna(df['avg_rating_of_driver'].sum()*1.0/len(df['avg_rating_of_driver']))
```

```python
df['phone'] = df['phone'].fillna('others')
```

```python
df = pd.concat((df, pd.get_dummies(df['phone']), pd.get_dummies(df['city']),pd.get_dummies(df['luxury_car_user'])), axis=1)
```

```python
df.head()
```

```python
df['churn'] = datetime(2014, 7, 1) - df['last_trip_date'] <= timedelta(days=30)
```

```python
df.info()
```

```python
df.rename(columns={'True': 'Luxury_car_user', 'False': 'Non_luxury_car_user'}, inplace=True)
```
